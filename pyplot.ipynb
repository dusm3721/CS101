```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.ylabel("some numbers")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], "ro")
plt.plot([1,2,3,4],[1,2,3,4], "bs")
plt.axis([0,5,0,20])
plt.show()
```

```python
x = np.arange(0.1, 4*np.pi, 0.1)
plt.plot(x, np.cos(x), "b-", x, np.sin(x), "r--", x, np.log(x), "g-.")
plt.show()
```

```python
data = {"x" : np.arange(50.0),
        "c" : np.random.randint(0,50,50),
        "s" : np.abs(np.random.randn(50)*100)
       }
data["y"] = data["x"] + 10 * np.random.randn(50)
plt.scatter(data["x"], data["y"], c=data["c"], s=data["s"])
plt.show()
```

```python
plt.subplot(211)
plt.plot(x, np.sin(x), "g-")

plt.subplot(212)
plt.plot(x, np.cos(x), "r-")
plt.show()
```

```python
y = np.random.randn(10000)
plt.figure(figsize=(10,7))
plt.hist(y, 50, normed=1, alpha=0.75)
plt.xlabel("x")
plt.ylabel("frequency")
plt.title("normal distribution")
plt.text(-2.9,0.403, "$\mu=0, \sigma=1$")
plt.grid(True)
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops={"facecolor" : 'black'},
             )

plt.ylim(-2, 2)
plt.show()
```
