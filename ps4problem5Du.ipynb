# Problem 5 Parametric VS Nonparametric Bootstrap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### (b)
From the distribution plots, we find that the parametric bootstrap gives a smoother distribution than the nonparametric one.

```python
r = 0
n, theta = 50, 1
np.random.seed(r)
sample = np.random.uniform(0, theta, n)
theta_hat = sample.max()

def f_theta_hat(x):
    if x>theta or x<0:
        return 0
    else:
        return n/theta**n * x**(n-1) 
```

```python
B = 10000
theta_hat_NP = []
np.random.seed(r)
for b in range(B):
    theta_hat_NP.append(np.random.choice(sample, size=n, replace=True).max())
```

```python
theta_hat_P = []
np.random.seed(r)
for b in range(B):
    theta_hat_P.append(np.random.uniform(0,theta_hat,n).max())
```

```python
x = np.arange(0,1.01,0.01)
plt.plot(x, [f_theta_hat(xi) for xi in x], color='r',label="theoretical")
plt.hist(theta_hat_NP, bins=40, normed=True,color='c', alpha=0.65,label="nonparametric")
plt.hist(theta_hat_P, bins=40, normed=True,color='y', alpha=0.65,label="parametric")
plt.legend()
plt.axis([0,1,0,150])
plt.xlabel("$\hat\Theta$")
plt.ylabel("distribution")
plt.grid(True)
plt.show()
```

```python
x = np.arange(0.90,1.002,0.002)
plt.plot(x, [f_theta_hat(xi) for xi in x], color='r',label="theoretical")
plt.hist(theta_hat_NP, bins=40, normed=True,color='c', edgecolor='k', alpha=0.65,label="nonparametric")
plt.hist(theta_hat_P, bins=40, normed=True,color='y', edgecolor='k', alpha=0.65,label="parametric")
plt.legend()
plt.axis([0.9,1,0,150])
plt.xlabel("$\hat\Theta$")
plt.ylabel("distribution")
plt.grid(True)
plt.show()
```
