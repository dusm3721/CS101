```python
ts = pd.Series(np.random.randn((1000)), index=pd.date_range("20120101", periods=1000))
ts = ts.cumsum()
ts.plot()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(1000,3), index=ts.index, columns=["A","B","C"])
df = df.cumsum()
df.plot()
plt.show()
```
